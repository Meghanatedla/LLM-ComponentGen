# Process the Test Results
## Imports and Helpers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

---

## Reading Data

```python
res = pd.read_csv('All Test Results.csv')
res.head()
```

---
## Processing Data

```python
res["Total Passed Before Percentage"] = (res["Total Passed Before"] / (res["Total Passed Before"] + res["Total Failed Before"])) * 100
res["Total Passed After Percentage"] = (res["Total Passed After"] / (res["Total Passed After"] + res["Total Failed After"])) * 100
res["Function Passed Before Percentage"] = (res["Function Passed Before"] / (res["Function Passed Before"] + res["Function Failed Before"])) * 100
res["Function Passed After Percentage"] = (res["Function Passed After"] / (res["Function Passed After"] + res["Function Failed After"])) * 100

res["Human Intervention Repo Level"] = (res["Total Passed H"] / (res["Total Passed Before"] + res["Total Failed Before"])) * 100
res["Human Intervention Function Level"] = (res["Function Passes H"] / (res["Function Passed Before"] + res["Function Failed Before"])) * 100

res.head()
```

```python
res.drop(columns=["Total Passed H", "Function Passes H"], inplace=True)
res.dropna(inplace=True)
```

```python
res = res[['Model', 'Repo', 'Type', 'Function', 'Total Passed Before Percentage', 'Total Passed After Percentage', 'Function Passed Before Percentage', 'Function Passed After Percentage', 'Human Intervention Repo Level', 'Human Intervention Function Level']]
res.head(15)
```

```python
res = res.sort_values(by=['Model', 'Repo', 'Type', 'Function'], ascending=[True, False, True, True])
res.drop('Function', axis=1, inplace=True)
res1 = res.groupby(['Model', 'Repo', 'Type']).mean().reset_index()
res1.head(55)
```

```python
res2 = res1.rename(columns={
    'Total Passed Before Percentage': 'Repo Level - Before',
    'Total Passed After Percentage': 'Repo Level - After - WH',
    'Function Passed Before Percentage': 'Function Level - Before',
    'Function Passed After Percentage': 'Function Level - After - WH',
    'Human Intervention Repo Level': 'Repo Level - After - H',
    'Human Intervention Function Level': 'Function Level - After - H'
})
res2.head()
```

### Pivoting the Data

```python
res_pivot = res2.pivot_table(index=['Model', 'Repo'],
                                columns='Type',
                                values=[
                                    'Repo Level - Before',
                                    'Function Level - Before',
                                    'Repo Level - After - WH', 'Function Level - After - WH',
                                    'Repo Level - After - H', 'Function Level - After - H'
                                ])

res_pivot = res_pivot.sort_index(axis=1, level=[1, 0])
res_pivot.columns = [f'{lvl2} - {lvl1}' for lvl1, lvl2 in res_pivot.columns]
res_pivot = res_pivot.reset_index()
res_pivot = res_pivot.sort_values(by=['Model', 'Repo'], ascending=[True, False])
res_pivot.drop(columns=['type 2 - Repo Level - Before', 'type 3 - Repo Level - Before', 'type 2 - Function Level - Before', 'type 3 - Function Level - Before'], inplace=True)
res_pivot.head()
```

```python
res_pivot = res_pivot.rename(columns={
    'type 1 - Repo Level - Before': 'Repo Level - Before',
    'type 1 - Function Level - Before': 'Function Level - Before',
})
final = res_pivot[['Model', 'Repo', 'Repo Level - Before', 'Function Level - Before', 'type 1 - Repo Level - After - WH', 'type 1 - Repo Level - After - H', 'type 1 - Function Level - After - WH', 'type 1 - Function Level - After - H', 'type 2 - Repo Level - After - WH', 'type 2 - Repo Level - After - H', 'type 2 - Function Level - After - WH', 'type 2 - Function Level - After - H', 'type 3 - Repo Level - After - WH', 'type 3 - Repo Level - After - H', 'type 3 - Function Level - After - WH', 'type 3 - Function Level - After - H']]


final = final.reset_index(drop=True)
final = final.round(2)
final.head(20)
```

```python
final_df = final.rename(columns={
    'Repo Level - Before': 'Before - Repo Level',
    'Function Level - Before': 'Before - Function Level',
    'type 1 - Repo Level - After - WH': 'After - type 1 - Repo Level - WH',
    'type 1 - Repo Level - After - H': 'After - type 1 - Repo Level - H',
    'type 1 - Function Level - After - WH': 'After - type 1 - Function Level - WH',
    'type 1 - Function Level - After - H': 'After - type 1 - Function Level - H',
    'type 2 - Repo Level - After - WH': 'After - type 2 - Repo Level - WH',
    'type 2 - Repo Level - After - H': 'After - type 2 - Repo Level - H',
    'type 2 - Function Level - After - WH': 'After - type 2 - Function Level - WH',
    'type 2 - Function Level - After - H': 'After - type 2 - Function Level - H',
    'type 3 - Repo Level - After - WH': 'After - type 3 - Repo Level - WH',
    'type 3 - Repo Level - After - H': 'After - type 3 - Repo Level - H',
    'type 3 - Function Level - After - WH': 'After - type 3 - Function Level - WH',
    'type 3 - Function Level - After - H': 'After - type 3 - Function Level - H'
})

final_df.head()
```

```python
final_df.to_csv('Processed Test Results.csv', index=False)
```
